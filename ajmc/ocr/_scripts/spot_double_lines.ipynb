# Introduction

This notebook aims at detecting double line (segmentation errors) in ocr datasets by computing the mean deviation for each line height. It can be used both for ajmc and PoGreTra.

# Detect double lines in ajmc

```python
from ajmc. text_processing.ocr_classes import OcrCommentary
from ajmc.commons import variables
from pathlib import Path
import numpy as np
from PIL import Image


for comm_id in variables.ALL_COMMENTARY_IDS:
    try:
        ocr_output_dir = next((Path(variables.COMMS_DATA_DIR) / comm_id / 'ocr/runs/').glob('*tess_base'))/ 'outputs'
    except StopIteration:
        continue

    ocr = OcrCommentary.from_ajmc_data(str(ocr_output_dir))

    for gt_page in ocr.ocr_gt_pages:
        line_heights = np.array([l.bbox.height for l in gt_page.children.lines])
        mean_line_height = np.quantile(line_heights, 0.5).mean()

        for line in gt_page.children.lines:
            if line.bbox.height > 1.9 * mean_line_height: # and line.bbox.height < 2 * mean_line_height:
                print(gt_page.id)
                print(line.text)
                display(Image.fromarray(gt_page.image.crop(line.bbox.bbox).matrix))


```
